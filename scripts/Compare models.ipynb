```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
fn_crys = 'model_C1_crys.csv'
fn_nocrys = 'model_C1_nocrys.csv'

dfcrys = pd.read_csv(fn_crys)
dfnocrys = pd.read_csv(fn_nocrys)
```

```python
umcrys = dfcrys['um']
modelcrys = dfcrys['model']

umnocrys = dfnocrys['um']
modelnocrys = dfnocrys['model']
```

```python
plt.figure(figsize = (16,8))
plt.subplot(211)
plt.plot(umcrys,modelcrys,color='red',label='crys included in model (best fit 9% fors)',lw=0.5)
plt.plot(umnocrys,modelnocrys,color='blue',label='crys not included in model',lw=0.5)
plt.xlabel('Wavelength')
plt.ylabel('Flux Density (W cm-2)')
plt.yscale('log')
plt.legend()
plt.subplot(212)
plt.plot(umcrys,(modelnocrys - modelcrys)/modelnocrys*100,color='grey')
plt.xlabel('Wavelength')
plt.ylabel(r'$(F_{\rm model, nocrys} - F_{\rm model, crys})/F_{\rm model, nocrys} (\%)$')
plt.minorticks_on()
plt.gca().tick_params(which='both',direction='in')
plt.savefig('crystalline_vs_none_model_comparison.jpg',bbox_inches='tight',dpi=150)
```
